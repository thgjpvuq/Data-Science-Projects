```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
ttnc = pd.read_csv("titanic.csv")
```

```python
ttnc.head()
```

```python
ttnc['Pclass'].nunique()
```

```python
ttnc['Survived'].sum()
```

```python
ttnc_class = ttnc.groupby('Pclass')
ttnc_class
```

```python
for emb, ttnc_df in ttnc_class:
    print(emb)
    print(ttnc_df)
```

```python
ttnc_class.get_group(1)
```

###### Aggregate

```python
# total number of people that survived per class
ttnc_class.sum()
```

```python
# average age
ttnc_class.mean()
```

```python
# max paid for tickets
ttnc_class.max()
```

```python
ttnc_class["Fare"].agg({"Fare" : ['sum', 'max']})
```

###### Filter

```python
ttnc_class.filter(lambda x: x['Age'].mean() > 38.0)
```

```python
ttnc_class.filter(lambda x: x['Age'].mean() < 38.0)
```

```python
ttnc_class.filter(lambda x: x['Age'].mean() > 38.0)['Fare'].mean()
```

```python
ttnc_class.filter(lambda x: x['Age'].mean() < 38.0)['Fare'].mean()
```

###### Transform

```python
ttnc['avg_fare_class'] = ttnc.groupby('Pclass')['Fare'].transform(lambda x: x.mean())
```

```python
ttnc.head()
```

```python
ttnc["fare_above_avg"] = ttnc["avg_fare_class"] < ttnc["Fare"]
ttnc
```
